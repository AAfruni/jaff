GENERATE C++ CODE COMPATIBLE WITH MICROPHYSICS

```python
# Create builder and generate amrexastro code
from jaff.network import Network
from jaff.builder import Builder
import os

# Load the network
network = Network("../networks/GOW.dat")
builder = Builder(network)
builds_dir = builder.build(template="microphysics")

print("\nC++ microphysics code generated successfully!")
print(f"Generated files are located at {builds_dir}")
```

```python
from jaff import Codegen
from jaff.network import Network
import re

network = Network("../networks/GOW.dat")
network.read_aux_funcs("../networks/GOW.jfunc")
cg = Codegen(network=network, lang="cxx")
result =cg.get_indexed_rhs(use_cse=False)
dedt_expr = result["expressions"][-1]
dedt_expr.value = re.sub(r"nden\[\s*(\d+)\s*\]", r"nden(\1)", dedt_expr.value)

print(f"{dedt_expr.value}")
```
